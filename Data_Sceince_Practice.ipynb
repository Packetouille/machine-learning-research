```python
import pandas as pd
data = {
    'Name': ['Alice', 'Bob', 'Alice', 'David', 'Alice'],
    'Age': [25, 30, 28, 40, 35],
    'Gender': ['Female', 'Male', 'Female', 'Male', 'Female'],
    'City': ['New York', 'London', 'New York', 'Paris', 'New York']
}

df = pd.DataFrame(data)

print(f"Original DataFrame: \n{df}")
```

```python
# Remove duplicate rows based on the 'Name' column, keeping the last occurrence, reset the index of resulting DataFrame
# NOTE: There may be cases where you DO NOT want to reindex. Such as if the indices represent unique identifiers for
# participants in a study.

df_cleaned = df.drop_duplicates(subset=['Name'], keep='last', ignore_index=True)
print(f"\nDataFrame after removing duplicates (keeping last occurrence of 'Name'):\n{df_cleaned}")
```

```python
# Sample DataFrame with specific indices
data = {
    'ID': [101, 102, 101, 103, 101],
    'Name': ['Alice', 'Bob', 'Alice', 'David', 'Alice'],
    'Age': [25, 30, 25, 35, 25],
    'City': ['New York', 'Los Angeles', 'New York', 'Chicago', 'New York']
}

df = pd.DataFrame(data)

# Display the original DataFrame
print("Original DataFrame:")
print(df)

# Remove duplicate rows based on the 'Name' column, keeping the last occurrence
df_cleaned = df.drop_duplicates(subset=['Name'], keep='last')

# Display the DataFrame after removing duplicates without resetting the index
print("\nDataFrame after removing duplicates (keeping last occurrence of 'Name') without resetting index:")
print(df_cleaned)
```

```python
# Sample DataFrame with specific indices and some missing
data = {
    'ID': [101, None, 101, None, 101, 106],
    'Name': ['Alice', 'Bob', 'Alice', 'David', 'Alice', 'Michael'],
    'Age': [25, 30, 25, 35, 25, 28],
    'City': ['New York', 'Los Angeles', 'New York', 'Chicago', 'New York', 'Boston']
}

df = pd.DataFrame(data)

# Display the original DataFrame
print("Original DataFrame:")
print(df)

# Generate new unique IDs for rows with missing IDs.
max_existing_id = df['ID'].max()

# Calculate the new unique IDs starting from the next integer
new_id_sequence = range(int(max_existing_id) + 1, int(max_existing_id) + 1 + df['ID'].isna().sum())

print(f"new_id_sequence: {new_id_sequence}")

# Assign new unique IDs to rows with missing IDs
df.loc[df['ID'].isna(), 'ID'] = new_id_sequence


```
